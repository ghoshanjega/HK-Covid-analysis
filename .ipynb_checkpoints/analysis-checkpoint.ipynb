# Data sourcing

## Hong Kong Government data

Source : [data.gov.hk](https://data.gov.hk/en-data/dataset/hk-dh-chpsebcddr-novel-infectious-agent/resource/7148ecd3-11ae-4ad1-90cf-345d2a2ee7a0)

```python
HKGov_df = pd.read_csv('HkGovData.csv')
HKGov_df['date'] = pd.to_datetime(HKGov_df['Report date'])
HKGov_df['date'] = pd.to_datetime(HKGov_df['date'].dt.strftime('%d/%m/%Y'))
HKGov_df = HKGov_df.sort_values('Case no.')
Confirmed = HKGov_df['Confirmed/probable'] =="Confirmed"
Confirmed_cases = HKGov_df[Confirmed]
Confirmed_cases = Confirmed_cases.reset_index()
Death_cases = HKGov_df[HKGov_df['Hospitalised/Discharged/Deceased'] == 'Deceased']

Confirmed_cases_count = Confirmed_cases.groupby('date').size()
idx = pd.date_range('01-01-2020', '12-30-2020')
# Set dates of confirmed cases as index
Confirmed_cases_count.index = pd.DatetimeIndex(Confirmed_cases_count.index)
# Fill in missing dates
Confirmed_cases_count = Confirmed_cases_count.reindex(idx, fill_value=0)
Death_cases_count = Death_cases.groupby('Report date').size()
```

```python

# for index in Confirmed_cases_count.index:
gov_df = Confirmed_cases_count.to_frame()
gov_df = gov_df.rename(columns = {0:'new_confirmed'})
gov_df['date'] = gov_df.index

gov_df['new_deceased'] = 0

for index in gov_df.index:
    try:
        gov_df.loc[index,'new_deceased'] = Death_cases_count.loc[index.strftime("%d/%m/%Y")]
    except KeyError:
        pass

# Calc aggregates
gov_df['total_confirmed'] = 0
gov_df['total_deceased'] = 0
total_confirmed = 0
total_deceased = 0
for index in gov_df.index:
    total_confirmed += gov_df.loc[index,'new_confirmed']
    gov_df.loc[index,'total_confirmed'] = total_confirmed
    total_deceased += gov_df.loc[index,'new_deceased']
    gov_df.loc[index,'total_deceased'] = total_deceased
   
```

```python
fig, axes = plt.subplots(nrows=4,ncols=1,figsize=(15,20))
## Plot New cases
gov_df.plot(ax = axes[0], subplots=True,x='date',y='new_confirmed') 
axes[0].set_title('New cases')
axes[0].set_xlabel('date')
axes[0].set_ylabel('# of cases')


## Plot Total cases
gov_df.plot(ax = axes[1], subplots=True,x='date',y='total_confirmed',color='black')  
axes[1].set_title('total cases')
axes[1].set_xlabel('date')
axes[1].set_ylabel('# of cases')
# print(axes[1].get_xticks())

## Plot New deceased
gov_df.plot(ax = axes[2], subplots=True,x='date',y='new_deceased')
axes[2].set_title('new deaths')
axes[2].set_xlabel('date')
axes[2].set_ylabel('# of deaths')

## Plot Total deceased
gov_df.plot(ax = axes[3], subplots=True,x='date',y='total_deceased',color='black')
axes[3].set_title('total deaths')
axes[3].set_xlabel('date')
axes[3].set_ylabel('# of deaths')

fig.suptitle('Hong Kong Covid cases 2020\nData source : data.gov.hk', fontsize=16)
fig.tight_layout()
```

## Google open data

Source : [Google big data](https://cloud.google.com/blog/products/data-analytics/free-public-datasets-for-covid19)

```python
goo_df = pd.read_csv(
    "google-open-data-hk.csv",
    keep_default_na=False,
    na_values=[""],
)
goo_df.sort_values(by=['date'])
goo_df.new_confirmed = goo_df.new_confirmed.fillna(value=0)
goo_df.new_deceased = goo_df.new_deceased.fillna(value=0)

goo_df.total_confirmed.fillna(method='ffill', inplace=True)
goo_df.total_deceased.fillna(method='ffill', inplace=True)
# goo_df.total_confirmed = hk_data.total_confirmed.fillna(value=0)
goo_df.date = pd.to_datetime(goo_df.date)
```

```python
fig, axes = plt.subplots(nrows=4,ncols=1,figsize=(15,20))

## Plot New cases
goo_df.plot(ax = axes[0], subplots=True,x='date',y='new_confirmed') 
axes[0].set_title('New cases')
axes[0].set_xlabel('date)')
axes[0].set_ylabel('# of cases')


## Plot Total cases
goo_df.plot(ax = axes[1], subplots=True,x='date',y='total_confirmed',color='black') 
axes[1].set_title('total cases')
axes[1].set_xlabel('date)')
axes[1].set_ylabel('# of cases')

## Plot New deceased
goo_df.plot(ax = axes[2], subplots=True,x='date',y='new_deceased')
axes[2].set_title('new deaths')
axes[2].set_xlabel('date)')
axes[2].set_ylabel('# of deaths')

## Plot Total deceased
goo_df.plot(ax = axes[3], subplots=True,x='date',y='total_deceased',color='black')
axes[3].set_title('total deaths')
axes[3].set_xlabel('date')
axes[3].set_ylabel('# of deaths')

fig.suptitle('Hong Kong Covid cases 2020\nData source : free-public-dataset-for-covid19 (google)', fontsize=16)
fig.tight_layout()
```

### Do restrictions work?

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(15,15))

## Plot New cases highlighted by government imposed restrictions
goo_df.plot(ax = axes[0], subplots=True,x='date',y='new_confirmed') 
for index in goo_df.index:
    if goo_df.loc[index,'restrictions_on_gatherings'] == 3.0:
        axes[0].axvspan(goo_df.loc[index,'date'],goo_df.loc[index,'date']+datetime.timedelta(1), color='dimgrey', alpha=0.05)
    if goo_df.loc[index,'restrictions_on_gatherings'] == 4.0:
        axes[0].axvspan(goo_df.loc[index,'date'],goo_df.loc[index,'date']+datetime.timedelta(1), color='black', alpha=0.05)
    
axes[0].legend(['new_confirmed','gatherings less than 10 people','gatherings between 10-100 people'])
leg = axes[0].get_legend()
leg.legendHandles[1].set_color('black')
leg.legendHandles[2].set_color('dimgrey')
axes[0].set_title('New cases against restrictions')
axes[0].set_xlabel('date')
axes[0].set_ylabel('# of cases')

## Plot Mobility against restrictions
# goo_df.plot(ax = axes[1], subplots=True,x='date',y='mobility_retail_and_recreation') 
axes[1].fill_between(goo_df.date,0,goo_df.mobility_retail_and_recreation,color="purple",alpha=0.2, label="Retail and recreation")
axes[1].fill_between(goo_df.date,0,goo_df.mobility_transit_stations,color="blue",alpha=0.2,label="Transit")
axes[1].fill_between(goo_df.date,0,goo_df.mobility_workplaces,color="indigo",alpha=0.2,label="Work Places")
axes[1].fill_between(goo_df.date,0,goo_df.mobility_parks,color="violet",alpha=0.2,label="Parks")
axes[1].fill_between(goo_df.date,0,goo_df.mobility_grocery_and_pharmacy,color="aqua",alpha=0.3,label="Grocery")
# axes[1].fill_between(goo_df.date,0,goo_df.mobility_grocery_and_pharmacy,color="yellow",alpha=0.3,label="Residential")
axes[1].plot(goo_df.date,goo_df.new_confirmed,label="new_confirmed (raw count)")
axes[1].legend()
for index in goo_df.index:
    if goo_df.loc[index,'restrictions_on_gatherings'] == 3.0:
        axes[1].axvspan(goo_df.loc[index,'date'],goo_df.loc[index,'date']+datetime.timedelta(1), color='dimgrey', alpha=0.05, label="Restriction 1")
    if goo_df.loc[index,'restrictions_on_gatherings'] == 4.0:
        axes[1].axvspan(goo_df.loc[index,'date'],goo_df.loc[index,'date']+datetime.timedelta(1), color='black', alpha=0.05)

fig.canvas.draw()
axes[1].set_title('Change in mobility')
axes[1].set_xlabel('date')
axes[1].set_ylabel('% change')


plt.show()
```
